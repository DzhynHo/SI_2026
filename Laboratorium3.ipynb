```python
# Biblioteka numpy uЕ‚atwia pracД™ z macierzami i wielowymiarowymi tabelami.
import numpy as np
# Biblioteka Tensorflow posЕ‚uЕјy nam do zbudowanie modelu
import tensorflow as tf
# Biblioteka opencv-python - biblioteka typu open source do przetwazanie obrazu i uczenia maszyny
import cv2
#Wizualizacja danych.
import matplotlib.pyplot as plt
# TensorFlow zawiera dostawcД™ danych dla MNIST, z ktГіrego bД™dziemy korzystaД‡.
import tensorflow_datasets as tfds
```

```python
# tfds.load w rzeczywistoЕ›ci Е‚aduje zestaw danych(lub pobiera, a nastД™pnie Е‚aduje, jeЕ›li uЕјywasz go po raz pierwszy)
# w naszym przypadku interesuje nas MNIST; nazwa zbioru danych jest jedynym obowiД…zkowym argumentem
# istniejД… inne argumenty, ktГіre moЕјemy okreЕ›licz i ktГіre moga nam siД™ przydaД‡
# with_info - True dostarczy nam rГіwnieЕј krotkД™ zawierajД…cД… informacje o wersji, funkcjach, liczbie prГіbek
# wykorzystamy te informacje nieco poniЕјej i zapiszemy je w mnist_info
# as_supervised = True zaЕ‚aduje zestaw danych w strukturze 2-krotkowej (wejЕ›cie,
# alternatywnie, as_supervised - False, zwrГіci sЕ‚ownik
# oczywiЕ›cie wolimy, aby nasze dane wejЕ›ciowe i cele byЕ‚y oddzielone
mnist_dataset, mnist_info = tfds.load(name='mnist', with_info=True, as_supervised=True)

# Po zaЕ‚adowaniu zestawu danych moЕјemy Е‚atwo wyodrД™bniД‡ zestaw danych szkoleniowyc
# i testowych z utworzonymi referencjami
mnist_train, mnist_test = mnist_dataset['train'], mnist_dataset['test']

# DomyЕ›lnie TF ma zestawy danych treningowych i testowych, ale nie ma zestawГіw walidacyjnych
# dlatego musimy go samodzielnie podzieliД‡
# zaczynamy od zdefiniowania liczby prГіbek walidacyjnych jako % prГіbek trenujД…cych
# tutaj rГіwnieЕј uЕјywamy mnist_info (nie musimy liczyД‡ obserwacji)
num_validation_samples = 0.1 * mnist_info.splits['train'].num_examples

# sparsujmy to na liczbД™ caЕ‚kowitД…, poniewaЕј liczba zmiennoprzecinkowa moЕјe po drodze spowodowaД‡ bЕ‚Д…d
num_validation_samples = tf.cast(num_validation_samples, tf.int64)

# zapiszmy teЕј liczbД™ prГіbek testowych w dedykowanej zmiennej (zamiast uЕјywaД‡ mnist_info)
num_test_samples = mnist_info.splits['test'].num_examples

# jeszcze raz parsujemy na liczbД™ caЕ‚kowitД… (zamiast domyЕ›lnej liczby zmiennoprzecinkowej)
num_test_samples = tf.cast(num_test_samples, tf.int64)

```

```python
# naleЕјy znormalizowaД‡ nasze dane, aby wynik byЕ‚ bardziej stabilny numerycznie
# dane wejЕ›ciowe muszД… byД‡ w zakresie liczbowym od 0 do 1
# zdefiniujmy funkcjД™ o nazwie: scale, ktГіra pobierze obraz MNIST i jego etykietД™

def scale(image, label):
# potrzebujemy wartoЕ›ci zmiennoprzecinkowej float
   image = tf.cast(image, tf.float32)
# poniewaЕј moЕјliwe wartoЕ›ci dla wejЕ›Д‡ to od 0 do 255 (256 rГіЕјnych odcieni szaroЕ›ci)
# jeЕ›li podzielimy kaЕјdy element przez 255, otrzymamy poЕјД…dany wynik:
# wszystkie elementy bД™dД… znajdowaД‡ siД™ w przedziale od 0 do 1
   image /= 255.
   return image, label

# metoda .map () pozwala nam zastosowaД‡ niestandardowД… transformacjД™ do danego zbioru danych
# juЕј zdecydowaliЕ›my, Ејe otrzymamy dane walidacyjne z mnist_train, wiД™c
scaled_train_and_validation_data = mnist_train.map(scale)

# na koniec normalizujemy i grupujemy dane testowe
# aby miaЕ‚y takД… samД… wielkoЕ›Д‡ jak dane trenjД…ce i walidacyjne
# nie ma potrzeby ich mieszania, poniewaЕј nie bД™dziemy trenowaД‡ na danych testowych
# bД™dzie to pojedyncza partia rГіwna wielkoЕ›ci danych testowych
test_data = mnist_test.map(scale)
```

```python
# ten parametr BUFFER_SIZE jest tutaj w przypadkach, gdy mamy do czynienia z ogromnymi zbiorami danych
# wtedy nie moЕјemy przetasowaД‡ caЕ‚ego zestawu danych za jednym razem,
# poniewaЕј nie moЕјemy zmieЕ›ciД‡ go w pamiД™ci
# wiД™c zamiast tego TF przechowuje w pamiД™ci tylko BUFFER_SIZE prГіbki na raz i tasuje je
# if BUFFER_SIZE = 1 => Ејadne tasowanie nie nastД…pi
# if BUFFER_SIZE> = liczba prГіbek => tasowanie jest jednolite
# BUFFER_SIZE pomiД™dzy - optymalizacja obliczeniowa w celu przybliЕјenia jednolitego tasowania
BUFFER_SIZE = 1000

# jest Е‚atwo dostД™pna metoda shuffle i musimy tylko okreЕ›liД‡ rozmiar bufora
shuffled_train_and_validation_data = scaled_train_and_validation_data.shuffle(BUFFER_SIZE)

# po przeskalowaniu i przetasowaniu danych moЕјemy przystД…piД‡ do faktycznego wyodrД™bniania
# danych trenujД…cych i walidacyjnych
# nasze dane walidacyjne byЕ‚yby rГіwne 10% zbioru uczД…cego, ktГіry juЕј obliczyliЕ›my
# uЕјywamy metody .take(), aby pobraД‡ tyle prГіbek
# na koniec tworzymy partiД™ o wielkoЕ›ci rГіwnej caЕ‚kowitej liczbie prГіbek walidacyjnych
validation_data = shuffled_train_and_validation_data.take(num_validation_samples)

# podobnie train_data to wszystko inne, wiД™c pomijamy tyle prГіbek
# ile jest w zbiorze danych walidacyjnych
train_data = shuffled_train_and_validation_data.skip(num_validation_samples)
```

```python
# zmienna okreЕ›lajД…ca wielkoЕ›Д‡ partii podawanej podczas trenowania
BATCH_SIZE = 100

# tworzymy partie danych treningowych
# jest to bardzo pomocne podczas treningu, poniewaЕј moЕјД™my iterowaД‡ po rГіЕјnych partiach
train_data = train_data.batch(BATCH_SIZE)

validation_data = validation_data.batch(num_validation_samples)

test_data = test_data.batch(num_test_samples)

# pobieramy i iterujemy partiД™ walidacyjnД… (jest to jedyna partia)
# poniewaЕј as_supervized = True, mamy strukturД™ skЕ‚adajД…cД… siД™ z dwГіch krotek
validation_inputs, validation_targets = next(iter(validation_data))
print(validation_inputs.shape, validation_targets.shape)
```

```python
input_size = 784 # 28 x 28 pikseli
output_size = 10 # 10 rГіЕјnych cyfr

# UЕјyjemy tego samego rozmiaru ukrytej warstwy dla obu ukrytych warstw. Nie jest to konieczne.
hidden_layer_size = 50

#MODEL
model = tf.keras. Sequential([

# Pierwsza warstwa to warstwa wejЕ›ciowa.
# KaЕјda prГіbka ma wymiary 28x28x1 pikseli, dlatego jest to tensor rozmiaru 3.
# PoniewaЕј to nie jest jeszcze CNN, nie wiemy, jak wprowadziД‡ takie dane wejЕ›ciowe
# do naszej sieci, wiД™c musimy spЕ‚aszczyД‡ obrazy.
# istnieje wygodna warstwa вЂћFlattenвЂќ, ktГіra po prostu pobiera nasz tensor 28x28x1
# i porzД…dkuje go w (None,) lub (28x28x1,)=(784,)
# to pozwala nam faktycznie stworzyД‡ sieД‡ neuronowД… typu feed forward
   tf.keras.layers.Flatten(input_shape=(28, 28, 1)), # input layer, warstwa wejЕ›ciowa

# tf.keras.layers.Dense jest w zasadzie implementacjД… modelu liniowego:
# y = xw + b czyli output = activation(dot(input, weight) + bias)
# wymaga kilku argumentГіw, ale najwaЕјniejsze dla nas to hidden_layer_size i funkcja aktywacji
tf.keras.layers.Dense(hidden_layer_size, activation='relu'), # warstwa ukryta
tf.keras.layers.Dense(hidden_layer_size, activation='relu'), # warstwa ukryta

# ostatnia warstwa nie jest inna, po prostu upewniamy siД™,
# Ејe aktywujemy jД… za pomocД… softmax, ktГіra daje nam rozkЕ‚ad prawdopodobieЕ„stwa
# i ma rozmiar output_size
tf.keras.layers.Dense(output_size, activation='softmax' ) # output layer

])

# okreЕ›lamy na koЕ„cu funkcjД™ optymalizacji, ktГіrego chcielibyЕ›my uЕјyД‡,
# funkcja straty, # oraz metryki, ktГіre chcemy uzyskaД‡ w kaЕјdej epocy uczenia
model. compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy' ])
```

```python
# ustawiamy iloЕ›Д‡ epok
NUM_EPOCHS = 30

# ustaw mechanizm wczesnego zatrzymywania, ktГіry ochroni nasz model przed przetrenowaniem
# ustawmy patience=2, aby byД‡ nieco tolerancyjnym na losowe wzrosty strat walidacji
early_stopping = tf.keras.callbacks. EarlyStopping(patience=2)

# dopasowujemy model, okreЕ›lajД…c dane treningowe, Е‚Д…cznД… liczbД™ epok
# oraz dane walidacyjne, ktГіre wЕ‚aЕ›nie stworzyliЕ›my, w formacie: (INPUTS, TARGETS),

model.fit(train_data, # dane wejЕ›ciowe trenujД…ce
          epochs=NUM_EPOCHS, # maksymalna iloЕ›Д‡ epok gdyby wczesne zatrzymanie nie zadziaЕ‚aЕ‚o
          callbacks=[early_stopping], # mechanizm "early stopping" - zapobieganie przetrenowaniu
          validation_data=(validation_inputs, validation_targets), # dane walidacyjne
          verbose = 1 # sposГіb pokazania treningu modelu
           )
```

```python
test_loss, test_accuracy = model.evaluate(test_data)
# moЕјemy zastosowaД‡ estetyczne formatowanie
print('Test loss: {0:.2f}. Test accuracy: {1:.2f}%'.format(test_loss, test_accuracy*100.))
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

file_names = [f"{i}.png" for i in range(10)]

plt.figure(figsize=(15, 5))

for i, name in enumerate(file_names):
    path = f'/content/liczby/{name}'

    if os.path.exists(path):

        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        img_resized = cv2.resize(img, (28, 28))

        img_final = img_resized.astype('float32') / 255
        img_final = np.reshape(img_final, (1, 28, 28, 1))

        prediction = model.predict(img_final, verbose=0)
        result = np.argmax(prediction)

        plt.subplot(1, 10, i + 1)
        plt.imshow(img_resized, cmap='gray')
        plt.title(f"Plik: {name}\nWynik: {result}")
        plt.axis('off')
    else:
        print(f"Brak pliku: {name}")

plt.show()
```
